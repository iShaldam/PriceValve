```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

## Sample ARIMA

```python
data = [10,34,2,235,100,223,123,455,324,123]
df = pd.DataFrame(data, columns=["Sales"])
```

```python
model = ARIMA(df["Sales"], order=(1,1,1))
model_fit = model.fit()
```

```python
future_steps = 5 
forecast = model_fit.forecast(steps= future_steps)
```

```python
plt.plot( df["Sales"], label = "First ", color='blue')
plt.scatter(range(len(df)), df["Sales"], color='blue')

```

```python
plt.plot(range(len(df), len(df) + future_steps) ,  
         forecast, label='Forecast', linestyle="dashed", color='red')
plt.scatter(range(len(df), len(df)+future_steps), forecast, color='red')
```

```python
plt.plot(range(len(df), len(df) + future_steps) ,  
         forecast, label='Forecast', linestyle="dashed", color='red')
plt.scatter(range(len(df), len(df)+future_steps), forecast, color='red')

plt.plot( df["Sales"], label = "First ", color='blue')
plt.scatter(range(len(df)), df["Sales"], color='blue')

plt.show()
```
